```python
import pandas as pd
import numpy as np
```

```python
stock_data = pd.read_csv('World-Stock-Prices-Dataset.csv')
```

```python
stock_data.head()
```

```python
stock_data = stock_data.drop(['Dividends', 'Stock Splits', 'Capital Gains'], axis=1)
```

```python
stock_data.head()
```

```python
stock_data.isna().any().any()
```

```python
print(stock_data['Country'].unique())
```

```python
print(stock_data['Industry_Tag'].unique())
```

Split data into different dataframes (one per table)

```python
# ticker (PK), brand name, industry tag, country
stock_desc = pd.DataFrame(data={'ticker': stock_data['Ticker'], 'name': stock_data['Brand_Name'], 'industry_tag': stock_data['Industry_Tag'], 'country': stock_data['Country']})

# ticker, date, open, high, low, close, volume
stock_price = pd.DataFrame(data={'ticker': stock_data['Ticker'], 'date': stock_data['Date'], 'open': stock_data['Open'], 'high': stock_data['High'], 'low': stock_data['Low'], 'close': stock_data['Close'], 'volume': stock_data['Volume']})
```

```python
# Convert the column to datetime format (if not already done)
stock_price['date'] = pd.to_datetime(stock_price['date'], utc=True)

# # Extract just the date
stock_price['date'] = stock_price['date'].dt.date
```

```python
len(stock_desc["ticker"].unique()) == len(stock_desc["ticker"])
```

```python
# remove duplicates by ticker
stock_desc = stock_desc.drop_duplicates(subset=['ticker'])
```

```python
len(stock_desc["ticker"].unique()) == len(stock_desc["ticker"])
```

```python
print(len(stock_desc))
```

```python
# add an id column to the first attribute of stock_price
stock_price.insert(0, 'id', range(1, 1 + len(stock_price)))
```
