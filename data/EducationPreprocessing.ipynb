```python
import pandas as pd
import numpy as np
```

```python
education_data = pd.read_csv('world_bank_education_data(1).csv')
```

```python
education_data.head(20)
```

```python
education_data.isna().any().any()
```

```python
print(education_data['country_name'].unique())
```

```python
# separate df with country code and name
country = pd.DataFrame(data={'country_code': education_data['country_code'], 'country_name': education_data['country_name']})

# separate df with country code, indicator code, year, and value
education = pd.DataFrame(data={'country_code': education_data['country_code'], 'indicator_code': education_data['indicator_code'], 'year': education_data['year'], 'value': education_data['value']})

# separate df with indicator code and name
education_indicators = pd.DataFrame(data={'indicator_code': education_data['indicator_code'], 'indicator_name': education_data['indicator_name']})
```

```python
len(country["country_code"].unique()) == len(country["country_code"])
```

```python
country = country.drop_duplicates(subset=['country_code'])
```

```python
education_indicators = education_indicators.drop_duplicates(subset=['indicator_code'])
```

```python
# check if there is a unique grouping of country code, indicator code, and year for each row in education
education.groupby(['country_code', 'indicator_code', 'year']).size().sort_values(ascending=False)
```
